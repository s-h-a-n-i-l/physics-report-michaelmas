```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
DATA_DIR = Path('../data/initial set up intergrator test 6.2.1 and gain calc data')
CSV_PATTERN = 'initial set up intergrator test 6.2.1 and gain calc data_*.csv'

frames = []
for csv_path in sorted(DATA_DIR.glob(CSV_PATTERN)):
    df = (
        pd.read_csv(csv_path, skiprows=[1])
        .dropna(subset=['Time', 'Channel A', 'Channel B'])
    )
    frames.append(df)

if not frames:
    raise FileNotFoundError(f'No CSV files matching {CSV_PATTERN!r} in {DATA_DIR}')

data = pd.concat(frames, ignore_index=True)
data = data.sort_values('Time').reset_index(drop=True)
data.head()
```
